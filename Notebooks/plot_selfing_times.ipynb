# Plot Selfing Timescale Figure
The idea is to plot the timescale of selfing.
Vary different selfing rates.
Save figures as PDFs without boundaries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def selfing_prob(t, s=0.9):
    """Gives the Probability of healing at time t ago
    t: Time ago. Can be also a Numpy array
    s: Selfing rate"""
    c = (s / 2.0) ** t # Calculate the Selfing Probability
    return c

def plot_timescale_selfing():
    """Plot the Timescale of coalescence for various selfing rates."""
    s_vec = [0.5, 0.7, 0.85, 0.95]
    colors = ["orange", "firebrick", "indigo", "steelblue"]
    t_max = 8 # Time back (max)
    fs=24 # Fontsize Plot
    lw=2.5 # Linewidth Plot
    ms=10
    
    t_vec = np.arange(1, t_max)
    
    res = np.zeros((len(s_vec), len(t_vec))) # Axis 1: Selfing rates, Axis 2: Time Points
    
    # Generate all the Values for Plotting:
    for i, s in enumerate(s_vec):
        res[i,:] = selfing_prob(t_vec, s)
        
    # Do the Plot:
    plt.figure(figsize=(6, 6))
    for i in range(len(s_vec)):
        plt.plot(t_vec, res[i,:], c=colors[i], linestyle="--", marker="o", 
                 label=("s=" + str(s_vec[i])), linewidth=lw, ms=ms)
    
    
    
    plt.xlabel("Generations Back", fontsize=fs)
    plt.ylabel("Probability Coalesent", fontsize=fs)
    plt.xticks(fontsize=fs/2)
    plt.yticks(fontsize=fs/2)
    plt.legend(fontsize=fs)
    plt.savefig('Figures/selfing_chance.pdf', bbox_inches='tight', pad_inches=0)
    plt.show()
    

plot_timescale_selfing()
```
